```python
"""
Model testing and evaluation script for
CNN-BiLSTM-Attention BP estimation model
"""

import os
import glob
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from scipy import signal
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import MinMaxScaler

# ---------------------------------------------------------------------
# === CUSTOM ATTENTION LAYER (must match training)
# ---------------------------------------------------------------------
class AttentionLayer(tf.keras.layers.Layer):
    def __init__(self, units=128, **kwargs):
        super(AttentionLayer, self).__init__(**kwargs)
        self.units = units

    def build(self, input_shape):
        self.W = self.add_weight(name='attention_weights',
                                 shape=(input_shape[-1], self.units),
                                 initializer='random_normal',
                                 trainable=True)
        self.b = self.add_weight(name='attention_bias',
                                 shape=(self.units,),
                                 initializer='zeros',
                                 trainable=True)
        self.u = self.add_weight(name='attention_context',
                                 shape=(self.units,),
                                 initializer='random_normal',
                                 trainable=True)
        super(AttentionLayer, self).build(input_shape)

    def call(self, inputs):
        uit = tf.nn.tanh(tf.tensordot(inputs, self.W, axes=1) + self.b)
        ait = tf.tensordot(uit, self.u, axes=1)
        ait = tf.nn.softmax(ait, axis=1)
        ait = tf.expand_dims(ait, axis=-1)
        weighted_input = inputs * ait
        return tf.reduce_sum(weighted_input, axis=1)

    def get_config(self):
        config = super().get_config()
        config.update({"units": self.units})
        return config

# ---------------------------------------------------------------------
# === SIGNAL PROCESSING HELPERS
# ---------------------------------------------------------------------
def preprocess_ppg_signal(ppg_signal):
    time_axis = np.arange(len(ppg_signal))
    coeffs = np.polyfit(time_axis, ppg_signal, 1)
    trend = np.polyval(coeffs, time_axis)
    detrended_signal = ppg_signal - trend
    scaler = MinMaxScaler(feature_range=(-1, 1))
    return scaler.fit_transform(detrended_signal.reshape(-1, 1)).flatten()

def extract_windows(ppg_signal, abp_signal, window_size=8.192, overlap=0.75, sampling_rate=125):
    win_len = int(window_size * sampling_rate)
    step = int(win_len * (1 - overlap))
    ppg_windows, sbp_targets, dbp_targets = [], [], []

    for start in range(0, len(ppg_signal) - win_len + 1, step):
        end = start + win_len
        ppg_w = ppg_signal[start:end]
        abp_w = abp_signal[start:end]
        peaks, _ = signal.find_peaks(abp_w, height=np.mean(abp_w), distance=int(0.6 * sampling_rate))
        troughs, _ = signal.find_peaks(-abp_w, height=-np.mean(abp_w), distance=int(0.6 * sampling_rate))
        if len(peaks) > 0 and len(troughs) > 0:
            sbp, dbp = np.mean(abp_w[peaks]), np.mean(abp_w[troughs])
        else:
            sbp, dbp = np.max(abp_w), np.min(abp_w)
        ppg_windows.append(ppg_w)
        sbp_targets.append(sbp)
        dbp_targets.append(dbp)

    return np.array(ppg_windows), np.array(sbp_targets), np.array(dbp_targets)

# ---------------------------------------------------------------------
# === PATIENT LOADER
# ---------------------------------------------------------------------
def load_patient(npz_file):
    try:
        d = np.load(npz_file)
        return d['ppg'].astype(np.float32), d['abp'].astype(np.float32)
    except Exception as e:
        print(f"⚠️ Failed to load {npz_file}: {e}")
        return None, None

# ---------------------------------------------------------------------
# === TEST DATASET GENERATOR
# ---------------------------------------------------------------------
def prepare_patient_windows(patient_files):
    X, y_sbp, y_dbp = [], [], []
    for file in patient_files:
        ppg, abp = load_patient(file)
        if ppg is None or abp is None:
            continue
        if len(ppg) < 1024:
            continue
        ppg_p = preprocess_ppg_signal(ppg)
        pw, sbp, dbp = extract_windows(ppg_p, abp)
        if len(pw) > 0:
            X.extend(pw)
            y_sbp.extend(sbp)
            y_dbp.extend(dbp)
    X = np.array(X).reshape(-1, 1024, 1)
    return X, np.array(y_sbp), np.array(y_dbp)

# ---------------------------------------------------------------------
# === PLOTTING HELPERS
# ---------------------------------------------------------------------
def plot_regression(true, pred, title):
    r2 = r2_score(true, pred)
    plt.figure(figsize=(5,5))
    plt.scatter(true, pred, alpha=0.5)
    plt.plot([true.min(), true.max()], [true.min(), true.max()], 'r--')
    plt.xlabel("True")
    plt.ylabel("Predicted")
    plt.title(f"{title}\nR² = {r2:.3f}")
    plt.grid(True)
    plt.show()

def plot_bland_altman(true, pred, title):
    mean = (true + pred) / 2
    diff = true - pred
    md, sd = np.mean(diff), np.std(diff)
    plt.figure(figsize=(6,5))
    plt.scatter(mean, diff, alpha=0.5)
    plt.axhline(md, color='gray', linestyle='--')
    plt.axhline(md + 1.96*sd, color='r', linestyle='--')
    plt.axhline(md - 1.96*sd, color='r', linestyle='--')
    plt.xlabel("Mean of True and Predicted")
    plt.ylabel("Difference (True - Pred)")
    plt.title(f"{title}\nMean diff={md:.2f}, SD={sd:.2f}")
    plt.grid(True)
    plt.show()

# ---------------------------------------------------------------------
# === MAIN TESTING PIPELINE
# ---------------------------------------------------------------------
def main():
    MODEL_PATH = r"\best_model_fold1.h5"
    DATA_PATH = r"\preprocessed_patients"

    print("Loading model...")
    model = tf.keras.models.load_model(MODEL_PATH, custom_objects={"AttentionLayer": AttentionLayer})
    print("✅ Model loaded successfully!")

    all_patients = sorted(glob.glob(os.path.join(DATA_PATH, "*.npz")))
    if len(all_patients) == 0:
        print("❌ No NPZ files found.")
        return

    # 🔀 Shuffle patients before splitting
    np.random.seed(42)
    np.random.shuffle(all_patients)
    print(f"Found {len(all_patients)} patients after shuffling")

    # 🧩 Automatically split into 5 equal parts (handles uneven sizes safely)
    segments = np.array_split(all_patients, 5)
    results = []

    for i, seg_files in enumerate(segments):
        print(f"\n🧠 Testing segment {i+1}/5 ({len(seg_files)} patients)...")
        X_test, y_sbp, y_dbp = prepare_patient_windows(seg_files)
        if len(X_test) == 0:
            print("⚠️ No valid data in this segment.")
            continue

        preds = model.predict(X_test, verbose=1)
        sbp_pred, dbp_pred = preds[0].flatten(), preds[1].flatten()

        sbp_mae = mean_absolute_error(y_sbp, sbp_pred)
        sbp_rmse = mean_squared_error(y_sbp, sbp_pred, squared=False)
        sbp_r2 = r2_score(y_sbp, sbp_pred)
        dbp_mae = mean_absolute_error(y_dbp, dbp_pred)
        dbp_rmse = mean_squared_error(y_dbp, dbp_pred, squared=False)
        dbp_r2 = r2_score(y_dbp, dbp_pred)

        results.append({
            "Segment": i+1,
            "SBP_MAE": sbp_mae, "SBP_RMSE": sbp_rmse, "SBP_R2": sbp_r2,
            "DBP_MAE": dbp_mae, "DBP_RMSE": dbp_rmse, "DBP_R2": dbp_r2
        })

        print(f"Segment {i+1}: SBP MAE={sbp_mae:.2f}, RMSE={sbp_rmse:.2f}, R²={sbp_r2:.3f} | "
              f"DBP MAE={dbp_mae:.2f}, RMSE={dbp_rmse:.2f}, R²={dbp_r2:.3f}")

        plot_regression(y_sbp, sbp_pred, f"SBP Regression - Segment {i+1}")
        plot_bland_altman(y_sbp, sbp_pred, f"SBP Bland–Altman - Segment {i+1}")
        plot_regression(y_dbp, dbp_pred, f"DBP Regression - Segment {i+1}")
        plot_bland_altman(y_dbp, dbp_pred, f"DBP Bland–Altman - Segment {i+1}")

    # 📊 Summary table
    print("\n📈 Summary of all segments:")
    for res in results:
        print(res)

    print("\n✅ Testing completed for all 5 segments!")

if __name__ == "__main__":
    main()
```
